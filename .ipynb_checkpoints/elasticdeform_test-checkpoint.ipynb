```python
import numpy, imageio, elasticdeform
import matplotlib.pyplot as plt
import tifffile as tiff
import numpy as np

X = numpy.zeros((100, 100, 100))
X[0:10, :, :] = 1

# apply deformation with a random 3 x 3 grid
# X_deformed = elasticdeform.deform_random_grid(X, sigma=5, points=3)

sigma =  25  # Standard deviation of the Gaussian
displacement = np.random.normal(size=(3,100,100,100)) * sigma

X_deformed = elasticdeform.deform_grid(X, displacement=displacement, order=3)

# X_deformed is the deformed X array. It has a shape like a gaussian spread from the original X.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import tifffile as tiff
import skimage as sk

# Create a grid of x and y values
resolution = 100
size = 30
x = np.linspace(-size,  size,  resolution+1)
y = np.linspace(-size,  size,  resolution+1)
x, y = np.meshgrid(x, y)


z1 = (-0.05*x**2-0.05*y**2+30) # top

mask_z1 = z1 > 0
z1 = np.where(mask_z1, z1,0)

# Create a list of z arrays
z_arrays = [z1]  # Add more z arrays to the list as needed

# Create a figure with subplots
fig = plt.figure(figsize=(15,  15))

# Create a  3D subplot for each z array
ax = fig.add_subplot(2,  2, 1, projection='3d')

for i, z in enumerate(z_arrays):
    ax.plot_surface(x, y, z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.tight_layout()
plt.show()

d3_array = np.zeros(shape=(resolution, resolution, resolution))

def update_d3_array(z, d3_array):
    for i in range(z.shape[0]-1):
        for j in range(z.shape[1]-1):
            if not np.isnan(z[i][j]):
                d3_array[i][j][int(z[i][j]*2)] =  1

# Update d3_array with each z array
update_d3_array(z1, d3_array)
```

```python
def normalize_image(im, mult_factor = 1):
    norm_im = (im - im.min()) / (im.max()-im.min()) * mult_factor
    return norm_im
```

```python
X = np.zeros((400,400))
# X[:, 0:50] = 1




# Define the center and radius of the circle
center_row = 200
center_col = 400 // 2  # Centered horizontally
radius = 100

# Iterate through each pixel in the array
for i in range(400):
    for j in range(400):
        # Calculate the distance from the current pixel to the center of the circle
        distance = np.sqrt((i - center_row)**2 + (j - center_col)**2)
        
        # Check if the distance is within the radius of the circle
        if distance <= radius:
            X[i, j] = 2  # Set the pixel value to 2 within the circle

radius = 90
# Iterate through each pixel in the array
for i in range(400):
    for j in range(400):
        # Calculate the distance from the current pixel to the center of the circle
        distance = np.sqrt((i - center_row)**2 + (j - center_col)**2)
        
        # Check if the distance is within the radius of the circle
        if distance <= radius:
            X[i, j] = 0  # Set the pixel value to 2 within the circle

X[::20, :] = 1
X[:, ::20] = 1

## 2 displacements (second one)
displacement = np.zeros((2,5,5))

displacement[0, 1, 2] = 100
displacement[0, 3, 2] = -100
displacement[1, 2, 1] = 100
displacement[1, 2, 3] = -100

displacement[0, 1, 1] = 40
displacement[1, 1, 1] = 40

displacement[0, 1, 3] = 40
displacement[1, 1, 3] = -40

displacement[0, 3, 1] = -40
displacement[1, 3, 1] = 40

displacement[0, 3, 3] = -40
displacement[1, 3, 3] = -40


print(displacement)

X_deformed = elasticdeform.deform_grid(X, displacement)
# End 2 displacements

plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(X)
plt.subplot(122)
plt.imshow(X_deformed)
```
