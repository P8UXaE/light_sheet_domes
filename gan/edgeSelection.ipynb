```python
import skimage as sk
import numpy as np
import napari
import matplotlib.pyplot as plt
import os
from csbdeep.io import save_tiff_imagej_compatible
```

```python
image = sk.io.imread('./../domes/Dome103_processed_n2v2d.tif')
# image = np.transpose(image, axes=(1, 0, 2))
```

```python
viewer = napari.Viewer()
viewer.add_image(image)
```

```python
z = 199

plt.imshow(image[z])
```

```python
y1 = 100
y2 = y1+100

x1 = 420
x2 = x1+100

im = image[z, y1:y2, x1:x2]

plt.figure(figsize=(30,7))

plt.subplot(1,4,1)
plt.imshow(im)
```

# Save the image in good_edges

```python
good_image_route = './dataset/good_edges'

file_number = 1
while f"{file_number}.tif" in os.listdir(good_image_route):
    file_number += 1

crop_filename = f"{good_image_route}/{file_number}.tif"

print(crop_filename)
save_tiff_imagej_compatible(crop_filename, im, axes='YX')
```

# Save the image in bad_edges

```python
bad_edges_route = './dataset/bad_edges'

file_number = 1
while f"{file_number}.tif" in os.listdir(bad_edges_route):
    file_number += 1

crop_filename = f"{bad_edges_route}/{file_number}.tif"

print(crop_filename)
save_tiff_imagej_compatible(crop_filename, im, axes='YX')
```
