```python
from dottamine import dotim
import numpy as np
import skimage as sk
# import napari
import matplotlib.pyplot as plt
```

```python
image = sk.io.imread('domes/70_denoised.tif')
# image = sk.io.imread('domes/Nice Domes/70.tif')
image = np.transpose(image[3], axes=(2,0,1))
```

```python
plt.imshow(image[300])
```

```python
dots = dotim(image=image[300])
```

```python
dots = dotim(image=image[3], is2D=False)
dots.get_mask(times=3, binshape=(2,2,2), v='+')

# 17m 3.8s times=3, binshape=(2,2,2)
```

```python
y = 200
dots2D = dotim(image=image[y], is2D=True)

plt.figure(figsize=(20,20))
plt.subplot(131)
plt.imshow(image[y])
plt.subplot(132)
plt.imshow(dots2D.conv_deconv(times=6, binshape=(2,2)))
plt.subplot(133)
plt.imshow(dots2D.get_mask())
```
