```python
print(image.shape)
center = (400, 400)
radius = 350
y, x = np.ogrid[:image.shape[1], :image.shape[2]]
mask = ((x - center[1])**2 + (y-center[0])**2 <= radius**2)

result = np.where(mask, image[150], 0)
plt.imshow(result)
```
