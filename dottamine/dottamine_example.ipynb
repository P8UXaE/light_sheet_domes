```python
from dottamine import dotim
import skimage as sk
import matplotlib.pyplot as plt
```

```python
image = sk.io.imread('./../../ppujol/dottamine_example_figures/21_denoised-tp5-z246.tif')
```

```python
dots = dotim(image=image, is2D=True)
convdeconv = dots.conv_deconv(times=3, binshape=(2,2)) # There is no need to run this conv_deconv process to get the mask.
convdeconv_dilated = sk.morphology.binary_dilation(convdeconv, footprint=sk.morphology.disk(2))
mask = dots.get_mask(morph_close_k=30, morph_open_k=1)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(131)
plt.imshow(image, cmap='binary_r')
plt.subplot(132)
plt.imshow(convdeconv_dilated, cmap='binary')
plt.subplot(133)
plt.imshow(mask, cmap='binary')
```

# convdeconv sequence

```python
convdeconv1 = dots.conv_deconv(times=1, binshape=(2,2))
convdeconv2 = dots.conv_deconv(times=2, binshape=(2,2))
convdeconv3 = dots.conv_deconv(times=3, binshape=(2,2))

plt.figure(figsize=(10,5))
plt.subplot(131)
plt.imshow(convdeconv1, cmap='binary')
plt.subplot(132)
plt.imshow(convdeconv2, cmap='binary')
plt.subplot(133)
plt.imshow(convdeconv3, cmap='binary')
```

```python
plt.figure(figsize=(10,10))
plt.subplot(221)
process1, _, _ = dots.poolAndMax2D(image)
plt.imshow(process1, cmap='binary_r')
plt.subplot(222)
process2, _, _ = dots.poolAndMax2D(process1)
plt.imshow(process2, cmap='binary_r')
plt.subplot(223)
process3, _, _ = dots.poolAndMax2D(process2)
plt.imshow(process3, cmap='binary_r')
plt.subplot(224)
process4, _, _ = dots.poolAndMax2D(process3)
plt.imshow(process4, cmap='binary_r')
```

```python
plt.figure(figsize=(12, 20))
plt.subplot(321)
plt.imshow(image, cmap='binary_r')
plt.subplot(323)
convdeconv1 = dots.conv_deconv(times=1, binshape=(1,1))
plt.imshow(convdeconv1, cmap='binary')
plt.subplot(325)
mask = dots.get_mask()
plt.imshow(mask, cmap='binary')


plt.subplot(322)
process1, _, _ = dots.poolAndMax2D(image)
plt.imshow(process1, cmap='binary_r')
plt.subplot(324)
convdeconv1 = dots.conv_deconv(times=1, binshape=(2,2))
plt.imshow(convdeconv1, cmap='binary')
plt.subplot(326)
mask = dots.get_mask(morph_close_k=30, morph_open_k=1)
plt.imshow(mask, cmap='binary')

plt.savefig('./../../ppujol/dottamine_example_figures/first.svg')
```
