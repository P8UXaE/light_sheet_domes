```python
from PIL import Image

# Initialize a black background image with the same size as your PNG images
background = Image.new('RGB', (1430, 913), (0, 0, 0))

# Iterate through each PNG image and composite it onto the black background
for i in range(36):
    # Load the transparent PNG image
    png_image = Image.open(f'./../../meshlab/stretched_snaps/snapshot{i:02d}.png')

    # Composite the PNG image onto the black background
    background.paste(png_image, (0, 0), png_image)

# Save the composite image as a TIFF file
background.save('./../../meshlab/stretched_map.tiff')
```

```python
import numpy as np
from PIL import Image

# Initialize an empty list to store image data
image_data = []

background = Image.new('RGB', (1430, 913), (0, 0, 0))

# Load each PNG image and append it to the list
for i in range(36):
    png_image = Image.open(f'./../../meshlab/stretched_snaps/snapshot{i:02d}.png')
    image_data.append(np.array(png_image))

# Stack the images along the third dimension to create a 3D array
# image_data = np.array(image_data)
print(image_data)
stacked_image = np.stack(image_data, axis=2)

# Convert the stacked image to uint8 format (assuming PNG images are 8-bit) 
stacked_image_uint8 = stacked_image.astype(np.uint8)

# Create an image object from the stacked image data
stacked_image_tiff = Image.fromarray(stacked_image_uint8)

# Save the stacked image as a TIFF file
stacked_image_tiff.save('./../../meshlab/stretched_map.tiff')
```
