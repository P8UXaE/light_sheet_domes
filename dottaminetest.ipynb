```python
from dottamine import dotim
import numpy as np
import matplotlib.pyplot as plt
import skimage as sk
```

```python
image = sk.io.imread('./../../Desktop/21_denoised-tp8.tif')
```

```python
y = 253
plt.imshow(image[y])
```

```python
dot1 = dotim(image=image, is2D=False)
dots1 = dot1.conv_deconv(times=3, binshape=(2,2,2), v='+')
```

```python
dot2 = dotim(image=image, is2D=False)
dots2 = dot2.conv_deconv(times=1, binshape=(8,8,8), v='+')
```

```python
plt.figure(figsize=(20,20))
plt.subplot(121)
plt.imshow(dots1[y])
plt.subplot(122)
plt.imshow(dots2[y])
```

```python
print(dots1.sum())
print(dots2.sum())
```

```python
import tifffile as tiff
tiff.imsave('./../../Desktop/dots1.tif', dots1)
tiff.imsave('./../../Desktop/dots2.tif', dots2)
```
