```python
from dottamine import dotim
import numpy as np
import skimage as sk
# import napari
import matplotlib.pyplot as plt
```

```python
image = sk.io.imread('domes/70_denoised.tif')
# image = sk.io.imread('domes/Nice Domes/70.tif')
image = np.transpose(image[3], axes=(2,0,1))
```

```python
# viewer = napari.Viewer()
# viewer.add_image(image)
```

```python
dotter = dotim(image=image, is2D=False)
mask = dotter.get_mask(times=3, binshape=(4, 4, 4), v='+')
# 35m 51.9s - times = 3 binshape = (2,2,2)
```

```python
im = image[409]
times = 3
binshape = (2,2)
dots2D = dotim(image=im, is2D=True)
plt.figure(figsize=(20,20))
plt.subplot(131)
plt.imshow(im)
plt.subplot(132)
plt.imshow(dots2D.conv_deconv(times=times, binshape=binshape))
plt.subplot(133)
plt.imshow(dots2D.get_mask())
```
