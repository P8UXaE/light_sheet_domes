```python
import skimage as sk
import matplotlib.pyplot as plt
import napari
import numpy as np
```

```python
image = sk.io.imread('./domes/domes_n2v/70_denoised.tif')
```

```python
viewer = napari.Viewer()
viewer.add_image(image)
```

```python
t = 3

z1 = 310
z2 = 385
z1_5 = int(np.array([z1, z2]).mean())

y1 = 380
y2 = 480
x1 = 250
x2 = 330

plt.figure(figsize=(40,20))

plt.subplot(1,3,1)
plt.imshow(image[t][z1])
plt.axhline(y1, c='r')
plt.axhline(y2, c='r')
plt.axvline(x1, c='r')
plt.axvline(x2, c='r')

plt.subplot(1,3,2)
plt.imshow(image[t][z1_5])
plt.axhline(y1, c='r')
plt.axhline(y2, c='r')
plt.axvline(x1, c='r')
plt.axvline(x2, c='r')

plt.subplot(1,3,3)
plt.imshow(image[t][z2])
plt.axhline(y1, c='r')
plt.axhline(y2, c='r')
plt.axvline(x1, c='r')
plt.axvline(x2, c='r')
```

```python
viewer.add_image(image[t][z1:z2, y1:y2, x1:x2])
```
